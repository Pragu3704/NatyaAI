```python
import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive
import os



# MediaPipe initialization
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
holistic = mp_holistic.Holistic(static_image_mode=False,
                                model_complexity=2,
                                enable_segmentation=True,
                                refine_face_landmarks=True)

def draw_enhanced_stick_figure(frame, results):
    h, w, _ = frame.shape
    canvas = np.zeros((h, w, 3), dtype=np.uint8)

    # Draw pose
    mp_drawing.draw_landmarks(
        canvas,
        results.pose_landmarks,
        mp_holistic.POSE_CONNECTIONS,
        mp_drawing.DrawingSpec(color=(255,255,255), thickness=2, circle_radius=2),
        mp_drawing.DrawingSpec(color=(255,255,255), thickness=2, circle_radius=2)
    )

    # Draw face
    if results.face_landmarks:
        mp_drawing.draw_landmarks(
            canvas,
            results.face_landmarks,
            mp_holistic.FACEMESH_TESSELATION,
            mp_drawing.DrawingSpec(color=(255,255,255), thickness=1, circle_radius=1),
            mp_drawing.DrawingSpec(color=(255,255,255), thickness=1, circle_radius=1)
        )

    # Draw hands
    if results.left_hand_landmarks:
        mp_drawing.draw_landmarks(
            canvas,
            results.left_hand_landmarks,
            mp_holistic.HAND_CONNECTIONS,
            mp_drawing.DrawingSpec(color=(255,255,255), thickness=2, circle_radius=2),
            mp_drawing.DrawingSpec(color=(255,255,255), thickness=2, circle_radius=2)
        )
    if results.right_hand_landmarks:
        mp_drawing.draw_landmarks(
            canvas,
            results.right_hand_landmarks,
            mp_holistic.HAND_CONNECTIONS,
            mp_drawing.DrawingSpec(color=(255,255,255), thickness=2, circle_radius=2),
            mp_drawing.DrawingSpec(color=(255,255,255), thickness=2, circle_radius=2)
        )

    return canvas

def process_video(input_path, output_path):
    cap = cv2.VideoCapture(input_path)

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Video writer setup
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Convert the BGR image to RGB
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Process the frame with MediaPipe
        results = holistic.process(rgb_frame)

        # Draw enhanced stick figure
        stick_figure = draw_enhanced_stick_figure(frame, results)

        # Write the frame
        out.write(stick_figure)

        # Display every 30th frame (adjust as needed)
        if frame_count % 30 == 0:
            cv2_imshow(stick_figure)
            print(f"Processing {os.path.basename(input_path)} - frame {frame_count}")

        frame_count += 1

    cap.release()
    out.release()
    print(f"Video processing complete. Output saved to: {output_path}")

# Specify the input and output folders
input_folder = '/content/drive/MyDrive/Dataset/'  # Update this path
output_folder = '/content/drive/MyDrive/Dataset/'  # Update this path

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Process all videos in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(('.mp4', '.avi', '.mov')):  # Add or remove video formats as needed
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, f'stick_figure_{filename}')
        process_video(input_path, output_path)

print("All videos processed.")
```

```python
# Step 1: Set up the environment and install MediaPipe
import os
import cv2
import mediapipe as mp
import numpy as np

# Step 2: Define a function to extract joint angles from a video
def extract_joint_angles(video_path):
    mp_pose = mp.solutions.pose
    pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5)

    cap = cv2.VideoCapture(video_path)
    angles = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(frame_rgb)

        if results.pose_landmarks:
            landmarks = results.pose_landmarks.landmark

            # Extract angles for all major joints (excluding head angles)
            angle_data = []

            # Left arm angles
            left_shoulder = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value]
            left_elbow = landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value]
            left_wrist = landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value]

            left_shoulder_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
            left_elbow_angle = calculate_angle(left_elbow, left_shoulder, left_wrist)
            angle_data.extend([left_shoulder_angle, left_elbow_angle])

            # Right arm angles
            right_shoulder = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value]
            right_elbow = landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value]
            right_wrist = landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value]

            right_shoulder_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
            right_elbow_angle = calculate_angle(right_elbow, right_shoulder, right_wrist)
            angle_data.extend([right_shoulder_angle, right_elbow_angle])

            # Left leg angles
            left_hip = landmarks[mp_pose.PoseLandmark.LEFT_HIP.value]
            left_knee = landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value]
            left_ankle = landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value]

            left_hip_angle = calculate_angle(left_hip, left_knee, left_ankle)
            left_knee_angle = calculate_angle(left_knee, left_hip, left_ankle)
            angle_data.extend([left_hip_angle, left_knee_angle])

            # Right leg angles
            right_hip = landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value]
            right_knee = landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value]
            right_ankle = landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value]

            right_hip_angle = calculate_angle(right_hip, right_knee, right_ankle)
            right_knee_angle = calculate_angle(right_knee, right_hip, right_ankle)
            angle_data.extend([right_hip_angle, right_knee_angle])

            # Append more joint angles as needed
            angles.append(angle_data)

    cap.release()
    pose.close()
    return np.array(angles)

# Helper function to calculate the angle between three points
def calculate_angle(a, b, c):
    a = np.array([a.x, a.y])  # First point
    b = np.array([b.x, b.y])  # Mid point
    c = np.array([c.x, c.y])  # End point

    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)

    if angle > 180.0:
        angle = 360.0 - angle

    return angle

# Step 3: Process the dataset and save the results
input_folder = '/content/drive/MyDrive/Dataset/Numpy'
output_folder = '/content/drive/MyDrive/Dataset/Numpy/Angles'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for video_file in os.listdir(input_folder):
    video_path = os.path.join(input_folder, video_file)
    angles = extract_joint_angles(video_path)

    # Save the angles to a .npy file
    output_path = os.path.join(output_folder, f"{os.path.splitext(video_file)[0]}_angles.npy")
    np.save(output_path, angles)
    print(f"Processed and saved: {video_file}")

print("All videos processed.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Load the saved angles from the .npy file
angles_file = '/content/drive/MyDrive/Dataset/Numpy/Angles/output_stick_figure_N11_angles.npy'
angles = np.load(angles_file)

print("Shape of angles array:", angles.shape)
print("First few rows of angles array:")
print(angles[:5])  # Print first 5 rows

# Define the human figure's initial pose in 3D
initial_pose = {
    'head': np.array([0, 0, 2]),
    'neck': np.array([0, 0, 1.8]),
    'left_shoulder': np.array([-0.2, 0, 1.8]),
    'left_elbow': np.array([-0.4, 0, 1.5]),
    'left_wrist': np.array([-0.6, 0, 1.2]),
    'right_shoulder': np.array([0.2, 0, 1.8]),
    'right_elbow': np.array([0.4, 0, 1.5]),
    'right_wrist': np.array([0.6, 0, 1.2]),
    'torso_center': np.array([0, 0, 1]),
    'left_hip': np.array([-0.2, 0, 0.8]),
    'left_knee': np.array([-0.3, 0, 0.4]),
    'left_ankle': np.array([-0.3, 0, 0]),
    'right_hip': np.array([0.2, 0, 0.8]),
    'right_knee': np.array([0.3, 0, 0.4]),
    'right_ankle': np.array([0.3, 0, 0])
}

def rotation_matrix(angle, axis):
    """Create a rotation matrix for the given angle (in degrees) around the given axis."""
    angle = np.radians(angle)
    c, s = np.cos(angle), np.sin(angle)
    if axis == 'x':
        return np.array([[1, 0, 0], [0, c, -s], [0, s, c]])
    elif axis == 'y':
        return np.array([[c, 0, s], [0, 1, 0], [-s, 0, c]])
    elif axis == 'z':
        return np.array([[c, -s, 0], [s, c, 0], [0, 0, 1]])

def apply_rotation(point, origin, rotation_matrix):
    """Apply rotation to a point around an origin."""
    return origin + np.dot(rotation_matrix, point - origin)

def update_pose_3d(pose, frame_angles):
    new_pose = pose.copy()

    # Left arm
    left_shoulder_rot = np.dot(rotation_matrix(frame_angles[1], 'x'), rotation_matrix(frame_angles[0], 'z'))
    new_pose['left_elbow'] = apply_rotation(pose['left_elbow'], pose['left_shoulder'], left_shoulder_rot)
    new_pose['left_wrist'] = apply_rotation(pose['left_wrist'], pose['left_shoulder'], left_shoulder_rot)

    left_elbow_rot = np.dot(rotation_matrix(frame_angles[1], 'x'), rotation_matrix(frame_angles[0], 'z'))
    new_pose['left_wrist'] = apply_rotation(new_pose['left_wrist'], new_pose['left_elbow'], left_elbow_rot)

    # Right arm
    right_shoulder_rot = np.dot(rotation_matrix(frame_angles[3], 'x'), rotation_matrix(frame_angles[2], 'z'))
    new_pose['right_elbow'] = apply_rotation(pose['right_elbow'], pose['right_shoulder'], right_shoulder_rot)
    new_pose['right_wrist'] = apply_rotation(pose['right_wrist'], pose['right_shoulder'], right_shoulder_rot)

    right_elbow_rot = np.dot(rotation_matrix(frame_angles[3], 'x'), rotation_matrix(frame_angles[2], 'z'))
    new_pose['right_wrist'] = apply_rotation(new_pose['right_wrist'], new_pose['right_elbow'], right_elbow_rot)

    # Left leg
    left_hip_rot = np.dot(rotation_matrix(frame_angles[5], 'x'), rotation_matrix(frame_angles[4], 'z'))
    new_pose['left_knee'] = apply_rotation(pose['left_knee'], pose['left_hip'], left_hip_rot)
    new_pose['left_ankle'] = apply_rotation(pose['left_ankle'], pose['left_hip'], left_hip_rot)

    left_knee_rot = np.dot(rotation_matrix(frame_angles[5], 'x'), rotation_matrix(frame_angles[4], 'z'))
    new_pose['left_ankle'] = apply_rotation(new_pose['left_ankle'], new_pose['left_knee'], left_knee_rot)

    # Right leg
    right_hip_rot = np.dot(rotation_matrix(frame_angles[7], 'x'), rotation_matrix(frame_angles[6], 'z'))
    new_pose['right_knee'] = apply_rotation(pose['right_knee'], pose['right_hip'], right_hip_rot)
    new_pose['right_ankle'] = apply_rotation(pose['right_ankle'], pose['right_hip'], right_hip_rot)

    right_knee_rot = np.dot(rotation_matrix(frame_angles[7], 'x'), rotation_matrix(frame_angles[6], 'z'))
    new_pose['right_ankle'] = apply_rotation(new_pose['right_ankle'], new_pose['right_knee'], right_knee_rot)

    return new_pose

# The rest of the code (figure setup, animation function, etc.) remains the same as in your original script

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

lines = {}
connections = [
    ('neck', 'left_shoulder'), ('left_shoulder', 'left_elbow'),
    ('left_elbow', 'left_wrist'), ('neck', 'right_shoulder'),
    ('right_shoulder', 'right_elbow'), ('right_elbow', 'right_wrist'),
    ('neck', 'torso_center'), ('torso_center', 'left_hip'),
    ('left_hip', 'left_knee'), ('left_knee', 'left_ankle'),
    ('torso_center', 'right_hip'), ('right_hip', 'right_knee'),
    ('right_knee', 'right_ankle')
]

for connection in connections:
    lines[connection], = ax.plot([], [], [], 'k-', lw=2)

def update(frame):
    current_pose = update_pose_3d(initial_pose, angles[frame])

    for connection, line in lines.items():
        start, end = connection
        line.set_data([current_pose[start][0], current_pose[end][0]],
                      [current_pose[start][1], current_pose[end][1]])
        line.set_3d_properties([current_pose[start][2], current_pose[end][2]])

    return lines.values()

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(0, 2)
ax.set_box_aspect((2, 2, 2))

anim = FuncAnimation(fig, update, frames=len(angles), interval=50, blit=False)

plt.show()
```
